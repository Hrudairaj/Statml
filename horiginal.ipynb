```python
#copy path
import pandas as pd
a=pd.read_csv('/content/bank.csv')
print(a)
```

```python
#y=a['job','education','default','loan','contact','month','poutcome']
x1=a.drop("job",axis=1)
x2=x1.drop("education",axis=1)
x3=x2.drop("default",axis=1)
x4=x3.drop("contact",axis=1)
x5=x4.drop("month",axis=1)
x6=x5.drop("poutcome",axis=1)
x7=x6.drop("loan",axis=1)
x8=x7.drop("marital",axis=1)
x9=x8.drop("deposit",axis=1)
x10=x9.drop("housing",axis=1)
print(x10)
#print(x2)
#print(x3)
```

```python
x=x10.iloc[:,0:6]
y=x10.iloc[:,6:7]
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
x_train, x_test, y_train, y_test= train_test_split(x , y, test_size=0.20, random_state=42)
print(y_train,x_train)
lg=LogisticRegression()
mm=lg.fit(x_train,y_train)
print(mm.score(x_test,y_test))
print(mm.score(x_train,y_train))
yp=mm.predict(x_test)
#print(accuracy_score(yp,y_test))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)

# Train a Logistic Regression model
lg = LogisticRegression()
mm = lg.fit(x_train, y_train)

# Print accuracy scores
print("Test Accuracy:", mm.score(x_test, y_test))
print("Train Accuracy:", mm.score(x_train, y_train))

# Predictions
yp = mm.predict(x_test)

# Create a confusion matrix
cm = confusion_matrix(y_test, yp)

# Plot the confusion matrix heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False,
            xticklabels=lg.classes_, yticklabels=lg.classes_)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
plt.show()

# Repeat the same process for an SVM model
from sklearn.svm import SVC

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)

# Train an SVM model
svm = SVC()
mm = svm.fit(x_train, y_train)

# Print accuracy scores
print("Test Accuracy:", mm.score(x_test, y_test))
print("Train Accuracy:", mm.score(x_train, y_train))

# Predictions
yp = mm.predict(x_test)

# Create a confusion matrix
cm = confusion_matrix(y_test, yp)

# Plot the confusion matrix heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False,
            xticklabels=svm.classes_, yticklabels=svm.classes_)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
plt.show()
```
